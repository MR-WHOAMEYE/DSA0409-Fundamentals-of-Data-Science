## 16. Scenario
You are working on a project that involves analyzing customer reviews for a product.  
You have a dataset containing customer reviews, and your task is to develop a Python program  
that calculates the frequency distribution of words in the reviews.

## Question
Develop a Python program to calculate the frequency distribution of words  
in the customer reviews dataset.


```python
import pandas as pd
import string
```

```python
reviews_data = pd.DataFrame({
    "review_id": [1, 2, 3],
    "review_text": [
        "This product is amazing! I really love the quality.",
        "Good product, but the delivery was slow.",
        "Amazing quality and great value for the price."
    ]
})
reviews_data
```

```python
all_reviews = " ".join(reviews_data["review_text"])
all_reviews = all_reviews.translate(str.maketrans("", "", string.punctuation))
words = all_reviews.split()
words
```

```python
word_frequency = {}
for word in words:
    if word in word_frequency:
        word_frequency[word] += 1
    else:
        word_frequency[word] = 1
for word, freq in word_frequency.items():
    print(f"{word}: {freq}")
```
