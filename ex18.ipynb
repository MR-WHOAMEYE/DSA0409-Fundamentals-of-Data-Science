## 18. Scenario
A hospital tested the age and body fat (%fat) data for 18 randomly selected adults.  
The dataset includes the age and corresponding body fat percentage for each individual.

## Question
1. Calculate the mean, median, and standard deviation of age and %fat using Pandas.  
2. Draw the boxplots for age and %fat.  
3. Draw a scatter plot and a Qâ€‘Q plot based on these two variables.

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
age = [23, 23, 27, 27, 39, 41, 47, 49, 50, 
       52, 54, 54, 56, 57, 58, 58, 60, 61]

fat = [9.5, 26.5, 7.8, 17.8, 31.4, 25.9, 27.4, 27.2, 31.2,
       34.6, 42.5, 28.8, 33.4, 30.2, 34.1, 32.9, 41.2, 35.7]
df = pd.DataFrame({"age": age,"%fat": fat})
df
```

```python
print("Mean:\n", df.mean())
print("\nMedian:\n", df.median())
print("\nStandard Deviation:\n", df.std())
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
df["age"].plot(kind="box", title="Boxplot of Age")
plt.subplot(1,2,2)
df["%fat"].plot(kind="box", title="Boxplot of %Fat")
plt.tight_layout()
plt.show()
```
